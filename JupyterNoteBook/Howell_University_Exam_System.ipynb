```python

#Assignment-2

totalscore=0
engflag=mathflag=gkflag=aptflag=True

def menu():
    print('1. Aptitude')
    print('2. English')
    print('3. Maths')
    print('4. G.K')
    print('5. Exit')
    print('\n')
    choice = int(input('Enter your choice between (1-5):'))
    return choice

def exam(ch):
        global aptflag,engflag,mathflag,gkflag,totalscore
        if ch==1:
            if aptflag:
                ans = input('How many weeks in a year is ?')
                if ans=='52':
                   totalscore = totalscore+10
                aptflag=False
            else:
                print('You have already attempted this subject...! Try another....')
            ch=menu()
            exam(ch)
        elif ch==2:
            if engflag:
                ans = input('How many vowels are in English aplhabets ?')
                if ans=='5':
                   totalscore = totalscore+10
                engflag=False
            else:
                print('You have already attempted this subject...! Try another....')
            ch=menu()
            exam(ch)
        elif ch==3:
            if mathflag:
                ans = input('Square root of 81 ?')
                if ans=='9':
                   totalscore = totalscore+10
                mathflag=False
            else:
                print('You have already attempted this subject...! Try another....')
            ch=menu()
            exam(ch)
        elif ch==4:
            if gkflag:
                ans = input('Name the bird which rotates its nect around 360 degree ?')
                if ans=='owl':
                   totalscore = totalscore+10
                gkflag=False
            else:
                print('You have already attempted this subject...! Try another....')
            ch=menu()
            exam(ch)
        elif ch==5:
            if totalscore ==20:
                bonus=2
            elif totalscore==30:
                bonus=5
            elif totalscore==40:
                bonus=10

            print('\n*************************************')
            print('Bonus points earned      :',bonus)
            finalscore = totalscore+bonus
            print('Total score out of 50 is :',finalscore)

            if finalscore>=40:
                print('You are a Genius')
            elif finalscore>=30:
                print('You are intelligent')
            elif finalscore>=20:
                print('IQ level is average')
            elif finalscore>=10:
                print('Your IQ level is below average')
            else:
                print('You need to re-appear the test')
            print('\n*************************************')

noatmpt = int(input('Enter no. of attempt :'))
if noatmpt > 1:
    print('You have already attempted this Exam.\nThank you....!')
else:
    choice = menu()
    exam(choice)
```
